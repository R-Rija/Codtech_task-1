#  Feature Understanding
(Univariate analysis)



```python
df['Year_Introduced'].value_counts()
```

```python
# Set a seaborn style
sns.set_style("whitegrid")

# Define a custom purple color palette
colors = sns.color_palette("Purples", 10)

ax = df['Year_Introduced'].value_counts() \
    .head(10) \
    .plot(kind='bar', 
          color=colors,   # Use the purple colors
          figsize=(10, 6), 
          title='Top 10 Years Coasters Introduced')

# Labels and title
ax.set_xlabel('Year Introduced', fontsize=12, weight='bold')
ax.set_ylabel('Count', fontsize=12, weight='bold')
ax.set_title('Top 10 Years Coasters Introduced', fontsize=14, weight='bold')

# Remove spines and show plot
sns.despine()
plt.tight_layout()
plt.show()

```

```python
# Use the same color palette for the histogram
custom_color = sns.color_palette("Purples", 1)[0]
#custom_color = sns.color_palette("Purples", 10)

ax = df['Speed_mph'].plot(kind='hist',
                          bins=20, 
                          color= custom_color,
                          figsize=(10, 6), 
                          title='Distribution of Coaster Speed (mph)')

ax.set_xlabel('Speed (mph)', fontsize=12, weight='bold')
ax.set_ylabel('Frequency', fontsize=12, weight='bold')
ax.set_title('Distribution of Coaster Speed (mph)', fontsize=14, weight='bold')

sns.despine()
plt.tight_layout()
plt.show()
```

```python
# Set a seaborn style for cleaner and more attractive visuals
sns.set_style("whitegrid")

# Define a custom color for the KDE plot
custom_color = sns.color_palette("coolwarm", 1)[0]

# Create a KDE plot for coaster speed
ax = df['Speed_mph'].plot(kind='kde',
                          color=custom_color,     # Use custom color
                          linewidth=2.5,          # Increase line thickness for better visibility
                          figsize=(10, 6),        # Set figure size
                          title='Density Plot of Coaster Speed (mph)')  # Updated title

# Set labels for x-axis and y-axis
ax.set_xlabel('Speed (mph)', fontsize=12, weight='bold')
ax.set_ylabel('Density', fontsize=12, weight='bold')

# Set a bold title with a bigger font size for emphasis
ax.set_title('Density Plot of Coaster Speed (mph)', fontsize=14, weight='bold')

# Add grid for better readability
ax.grid(True, linestyle='--', alpha=0.7)

# Remove top and right spines for a cleaner look
sns.despine()

# Show the plot
plt.tight_layout()  # Adjust layout to avoid overlap
plt.show()
```

#  Feature Relationships


```python
# Set a seaborn style for consistency
sns.set_style("whitegrid")

# Use a custom color palette for the scatter plot (you can also use a single color)
scatter_color = sns.color_palette("coolwarm", 1)[0]

# Create the scatter plot
ax = df.plot(kind='scatter',
             x='Speed_mph', 
             y='Height_ft',
             color=scatter_color,  # Consistent color with other plots
             figsize=(10, 6),      # Set figure size for consistency
             title='Coaster Speed vs. Height')

# Set labels for x and y axes
ax.set_xlabel('Speed (mph)', fontsize=12, weight='bold')
ax.set_ylabel('Height (ft)', fontsize=12, weight='bold')

# Set a bold title for emphasis
ax.set_title('Coaster Speed vs. Height', fontsize=14, weight='bold')

# Remove top and right spines for a cleaner look
sns.despine()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Set a seaborn style for consistency across all visualizations
sns.set_style("whitegrid")

# Use a color palette that matches the rest of the project for 'Year_Introduced'
palette = sns.color_palette("coolwarm", as_cmap=True)

# Create the scatter plot using seaborn
ax = sns.scatterplot(x='Speed_mph',
                     y='Height_ft',
                     hue='Year_Introduced',   # Different colors for 'Year_Introduced'
                     palette=palette,         # Use consistent color palette
                     data=df,
                     s=100,                   # Set marker size for better visibility
                     edgecolor='black',        # Add edge color for better definition
                     alpha=0.7)               # Add transparency for better overlapping points

# Set a bold title for the plot
ax.set_title('Coaster Speed vs. Height', fontsize=14, weight='bold')

# Set axis labels with bold font
ax.set_xlabel('Speed (mph)', fontsize=12, weight='bold')
ax.set_ylabel('Height (ft)', fontsize=12, weight='bold')

# Remove top and right spines for a cleaner look
sns.despine()

# Show the plot
plt.tight_layout()
plt.show()
```

# Step 5: Ask a Question about the data
- Try to answer a question you have about the data using a plot or statistic.

What are the locations with the fastest roller coasters (minimum of 10)?

```python
ax = df.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean','count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean'] \
    .plot(kind='barh', figsize=(12, 5), title='Average Coast Speed by Location')
ax.set_xlabel('Average Coaster Speed')
plt.show()
```
